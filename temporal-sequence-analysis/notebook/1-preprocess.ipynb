```python
import cv2
import numpy as np
import os
```

```python
def preprocess(oct_img):
    tomogram = oct_img[0:496, 496:]
    height, width, _ = tomogram.shape
    if height != width:
        new_size = max(width, height)
        new_image = np.full((new_size, new_size, 3), (0,0,0), dtype=np.uint8)
        left = (new_size - width) // 2
        top = (new_size - height) // 2
        new_image[top:top+height, left:left+width] = tomogram
        tomogram = new_image
    return cv2.resize(tomogram, (600, 600))
```

```python
scan_types = ["raster", "radial"]
for st in scan_types:
    input_dir = f"../data/original/{st}"
    output_dir = f"../data/preprocessed/{st}"
    for filename in os.listdir(input_dir):
        if filename.endswith(".png") or filename.endswith(".jpg"):
            img = cv2.imread(os.path.join(input_dir, filename))
            preprocessed_img = preprocess(img)
            cv2.imwrite(os.path.join(output_dir, filename), preprocessed_img)
```
