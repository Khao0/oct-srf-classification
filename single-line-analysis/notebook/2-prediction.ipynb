```python
import numpy as np
import cv2
import tensorflow as tf
```

```python
retinal_fundus = cv2.imread('../data/retinal_fundus.png')
tomogram = cv2.imread('../data/tomogram.png')
input_features = [
        np.expand_dims(retinal_fundus, axis=0),
        np.expand_dims(tomogram, axis=0)
]
```

```python
CLASS_NAMES = ["CSC", "Normal", "PCV", "VKH"]
```

```python
single_line_model = tf.keras.models.load_model("../../inference_service/models/single-line-model.h5")
single_line_model.trainable = False
```

```python
output = single_line_model.predict(input_features)

scores = output[0].tolist()
pred_class = CLASS_NAMES[np.argmax(scores)]
```

```python
print(f"Predicted class: {pred_class}")
print(f"Class scores: {scores}")
```
